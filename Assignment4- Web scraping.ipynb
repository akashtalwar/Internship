Q2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening url
url = "https://www.bcci.tv/"
driver.get(url)
time.sleep(2)

#Click on fixtures(at page end)
link = driver.find_element_by_xpath('/html/body/footer/div/nav/ul/li[2]/ul/li[1]/a')  
link.click()
time.sleep(2)

#Empty list
match=[]
series=[]
place=[]
date=[]
Time=[]

#Scrape match
matches=driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for i in matches:
    match.append(i.text)

#Scrape series
ser=driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]')
for i in ser:
    series.append(i.text)

#Scrape place
for i in range(1):
    places=driver.find_elements_by_xpath('/html/body/div[4]/div/div/div[2]/section/div/div/a["i"]/div[2]/div[2]/p/span')
    for i in places:
        place.append(i.text)
```

```python
date=[]
#Scrape date
for i in range(1):
    dates=driver.find_elements_by_xpath('/html/body/div[4]/div/div/div[2]/section/div/div/a["i"]/div[2]/div[2]/span/strong')
    for j in dates:
        date.append(j.text) 
```

```python
#Scrape time
tim=[]
Time=[]
for i in range(1):
    times=driver.find_elements_by_xpath('/html/body/div[4]/div/div/div[2]/section/div/div/a["i"]/div[2]/div[2]/span')
    for j in times:
        tim.append(j.text)
for i in range(0,11):
    Time.append(tim[i][-9:])
```

```python
#Create Dataframe
fixture=pd.DataFrame()
fixture["Match"]= pd.Series(match)
fixture["Series"]= pd.Series(series)
fixture["Place"]= pd.Series(place)
fixture["Date"]= pd.Series(date)
fixture["Time"]= pd.Series(Time)
fixture
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening url
url = "https://github.com/"
driver.get(url)
time.sleep(4)

#Click on "explore" tab
link1=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary')
link1.click()
time.sleep(3)

#Click on "trending" tab
link2=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a')
link2.click()
time.sleep(3)

#Repositary Title & Repositary description
name=[]
for i in range(1):
    names=driver.find_elements_by_xpath('/html/body/div[4]/main/div[3]/div/div[2]/article["i"]/h1/a')
    for j in names:
        name.append(j.text)

git_hub=pd.DataFrame()
git_hub["Name"]=pd.Series(name)  #Assigning to list to dataframe     
git_hub["Repositary Title"],git_hub["Repositary description"]=git_hub["Name"].str.split("/",2).str
git_hub.drop(columns=["Name"],inplace=True)

#Number of contrubutors
num=[]
for i in range(1):
    number=driver.find_elements_by_xpath('/html/body/div[4]/main/div[3]/div/div[2]/article["i"]/div[2]/a[2]')
    for j in number:
        num.append(j.text) 
git_hub["No. of Contributors"]=pd.Series(num) #Assigning to list to dataframe

#Language
lang=[]
for i in range(1):
    language=driver.find_elements_by_xpath('/html/body/div[4]/main/div[3]/div/div[2]/article["i"]/div[2]/span[1]/span[2]')
    for j in language:
        if j.text is not str(" "):
            lang.append(j.text)
        else:
            lang.append(np.nan)
lang.insert(4,"-")
lang.insert(14,"-")
lang.insert(17,"-")
lang.insert(22,"-")
git_hub["Language used"]=pd.Series(lang) #Assigning to list to dataframe
git_hub #Final dataframe
```

Q6. Scrape the details of top 100 songs on billboard.com.
Url = https://www.billboard.com/
You have to find the following details:
A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board
Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening url
url = "https://www.billboard.com/"
driver.get(url)
time.sleep(2)

#Click on "charts" tab
link1=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/header/div/ul/li[1]/a')
link1.click()
time.sleep(2)

#Click on "Hot 100" tab
link2=driver.find_element_by_xpath('/html/body/main/div[6]/div/div[2]/div[2]/a[1]')
link2.click()
time.sleep(2)

hot_100=pd.DataFrame() #Empty Dataframe

#Scrape song
song=[]
for i in range(1):
    songs=driver.find_elements_by_xpath('/html/body/main/div/div/div[7]/div/ol/li["i"]/button/span[2]/span[1]')
    for j in songs:
        song.append(j.text)
hot_100["Song_name"]=pd.Series(song)

#Scrape artist
artist=[]
for i in range(1):
    artists=driver.find_elements_by_xpath('/html/body/main/div/div/div[7]/div/ol/li["i"]/button/span[2]/span[2]')
    for j in artists:
        artist.append(j.text)
hot_100["Artist_name"]=pd.Series(artist)

#Scrape last week rank,Peak rank & Weeks on chart
rank=[]
for i in range(1):
    ranks=driver.find_elements_by_xpath('/html/body/main/div/div/div[7]/div/ol/li["i"]/button/div')
    for j in ranks:
        rank.append(j.text)
hot_100["Rank"]=pd.Series(rank)

#Split rank into 3 columns
hot_100["Last Week"],hot_100["Peak Rank"],hot_100["Weeks on Chart"]=hot_100["Rank"].str.split("\n",3).str
hot_100.drop(columns=["Rank"],inplace=True)

#Final dataframe
hot_100
```

Q7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/
You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening url
url = "https://www.naukri.com/"
driver.get(url)
time.sleep(2)
```

```python
window_before = driver.window_handles[0] #Parent window
#Click on "Recruiters" tab
driver.find_element_by_xpath('/html/body/div[1]/div[1]/div/ul[1]/li[2]/a/div').click()
time.sleep(2)
window_after = driver.window_handles[1] #New window
driver.switch_to_window(window_after) #Switch to new window
```

```python
#Enter "Data Science"
search=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/div/div[1]/div[1]/div[2]/input')
search.send_keys("Data science")     
time.sleep(2)
#Click on "search" tab
link2=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/button')
link2.click() 
```

```python
#Empty list
name=[]
designation=[]
company=[]
skill=[]
location=[]

#Scrape all data
data=[]
for i in range(1):
    design=driver.find_elements_by_xpath('//p[@class="highlightable"]')
    for j in design:                      
        data.append(j.text.split("\n"))

#Scrape name, designation,location
name=[]
designation=[]
company=[]
for i in range(0,50):
    name.append(data[i][0]) #
    designation.append(data[i][1])
    company.append(data[i][2])
        
#Scrape location
for i in range(1):
    loc=driver.find_elements_by_xpath('//small[@class="ellipsis"]')
    for j in loc:
        if j.text==" ":
            location.append("None")
        else:
            location.append(j.text)
location.insert(26,"None") #insert missinng values
location.insert(48,"None") #Insert missing values

#Scrape skills
for i in range(1):
    skills=driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
    for j in skills:
        skill.append(j.text)

#Create dataframe
naukri=pd.DataFrame()
naukri["Name"]=pd.Series(name)
naukri["Designation"]=pd.Series(designation)
naukri["Company"]=pd.Series(company)
naukri["Skill they hire"]=pd.Series(skill)
naukri["Location"]=pd.Series(location)
naukri
```

Q9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# opening url
url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(2)

#Empty List
Name=[]
years=[]
genre=[]
run_time=[]
rating=[]
vote=[]

for i in range(1):
    #Scrape Name
    Names=driver.find_elements_by_xpath('/html/body/div[4]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div["i"]/div[2]/h3/a')
    for j in Names:
        Name.append(j.text)
    
    #Scrape years
    years=driver.find_elements_by_xpath('/html/body/div[4]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div["i"]/div[2]/h3/span[2]')
    for j in years:
        year.append(j.text)
        
    #Scrape genre
    genres=driver.find_elements_by_xpath('/html/body/div[4]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div["i"]/div[2]/p[1]/span[5]')
    for j in genres:
        genre.append(j.text)
    
    #Scrape run_time
    times=driver.find_elements_by_xpath('/html/body/div[4]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div["i"]/div[2]/p[1]/span[3]')
    for j in times:
        run_time.append(j.text) 
    
    #Scrape rating
    ratings=driver.find_elements_by_xpath('/html/body/div[4]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div["i"]/div[2]/div[1]/div[1]/span[2]')
    for j in ratings:
        rating.append(j.text)  
    
    #Scrape vote
    votes=driver.find_elements_by_xpath('/html/body/div[4]/div/div[2]/div[3]/div[1]/div/div[3]/div[3]/div["i"]/div[2]/p[4]/span[2]')
    for j in votes:
        vote.append(j.text) 
        
#Empty dataframe
web_series=pd.DataFrame()
web_series["Name"]=pd.Series(Name)
web_series["Year Span"]=pd.Series(year)
web_series["Genre"]=pd.Series(genre)
web_series["Run_time"]=pd.Series(run_time)                        
web_series["Ratings"]=pd.Series(rating)                        
web_series["Vote"]=pd.Series(vote)
web_series
```
